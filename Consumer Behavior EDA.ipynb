Notes about the dataset:

- No missing values
- Customer ID's are unique values, this means the dataset comprehends only single item purchases.
- Customers previous purchases are counted but the total money spent remains unknown.
- Gender is separeted only in two categories. Those who idenfify as male represent 2/3 of the customer base.

```python
# Best products in different categories
cat_items = df[["Item Purchased", "Category", "Purchase Amount (USD)", "Color", "Season", "Review Rating"]]
cat_items
```

```python
# Count the number of purchased items per season
items_per_season = cat_items['Season'].value_counts()


items_per_season.plot(kind='bar')


plt.title('Number of Items Purchased per Season')
plt.xlabel('Season')
plt.ylabel('Number of Items Purchased')
```

```python
# Sums spend in categories. Add mean.
cat_sums = df.groupby("Category")["Purchase Amount (USD)"].sum().reset_index()
cat_sums
```

```python
# Best products per season
cat_items.groupby("Season")["Item Purchased"].value_counts().reset_index()
```

```python
# Best colors per season
cat_items.groupby("Season")["Color"].value_counts()
```

```python
# Highest rated products and lowest rated products
pd.crosstab(cat_items["Item Purchased"], cat_items["Review Rating"])
```

```python
# Group the data by 'Category' and then find the index of the max 'Review Rating' for each category
idx = cat_items.groupby('Category')['Review Rating'].idxmax()

# Use these indices to find the corresponding rows in the original dataframe
best_products_per_category = cat_items.loc[idx]

# Create a bar chart with 'Category' on the x-axis and 'Review Rating' on the y-axis
plt.figure(figsize=(10, 6))  # This sets the size of the figure, adjust as necessary
plt.bar(best_products_per_category['Category'], best_products_per_category['Review Rating'])

# Add labels and title
plt.xlabel('Category')
plt.ylabel('Review Rating')
plt.title('Best Rated Product per Category')

# If you want to show the product name as well, you can use 'plt.text' to add annotations
for i in range(len(best_products_per_category)):
    plt.text(i, best_products_per_category['Review Rating'].iloc[i] + 0.02, # This positions the text slightly above the bar
             best_products_per_category['Item Purchased'].iloc[i], 
             ha='center')  # 'ha' is the horizontal alignment
```

- Purchase Amount (USD) median is the same (60.0) between sexes and subscribed/non-subscribed customers.

- Higher number of previous purchases may predict higher spend per transaction.
